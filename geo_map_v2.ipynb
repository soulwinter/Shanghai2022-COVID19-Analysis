```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import pyecharts.charts as pyc

path = r'D:\COVID-19\Shanghai2022-COVID19-Analysis\DataInfo2\community\20220306_geo.xlsx'
```

```python
df = pd.read_excel(path)

df_quezhen = df[df["类型"].isin(["确诊"])]
df_wuzhengzhuang = df[df["类型"].isin(["无症状"])]
```

```python
from pyecharts import options as opts
from pyecharts.charts import Geo
from pyecharts.globals import GeoType
```

```python
# 绘制散点图

geo = Geo(is_ignore_nonexistent_coord=True)
geo.add_schema(maptype="上海")  # 使用上海地图

for i in df_quezhen.index:
    s = df_quezhen.loc[i]
    geo.add_coordinate(s["地址"], s["经度"], s["纬度"])

qz_data_pair = [(df_quezhen.loc[i]["地址"],0)for i in df_quezhen.index]

for i in df_wuzhengzhuang.index:
    s = df_wuzhengzhuang.loc[i]
    geo.add_coordinate(s["地址"], s["经度"], s["纬度"])

wzz_data_pair = [(df_wuzhengzhuang.loc[i]["地址"],1)for i in df_wuzhengzhuang.index]

geo.add("",qz_data_pair,type_ =GeoType.EFFECT_SCATTER )
geo.add("",wzz_data_pair,type_ =GeoType.EFFECT_SCATTER)

geo.set_series_opts(label_opts=opts.LabelOpts(is_show=False))  # 不展示数据点label
geo.set_global_opts(
    visualmap_opts=opts.VisualMapOpts(max_=550,
                                             is_piecewise=True,
                                             pieces=[
                     {"max": 0, "min": 0, "label": "确诊", "color": "#708090"},
                     {"max": 1, "min": 1, "label": "无症状", "color": "#FF0000"}]),
    title_opts=opts.TitleOpts(title="上海疫情散点图") 
    )
geo.render_notebook()
```
