# 新冠肺炎区域每日新增

## 包括两个数据：确证病例数和无症状感染者数

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import pyecharts.charts as pyc
```

```python
data = pd.read_csv('Shanghai_region_data_T.csv')
df = pd.DataFrame(data)
```

将确诊病例数与无症状感染者数分开

```python
df_quezhen = df.loc[:,~df.columns.str.contains('wuzhengzhuang')] 
df_wuzhengzhuang = df.loc[:,~df.columns.str.contains('quezhen')] 
```

获取日期列表和地区名列表，用于绘图

```python
date_list = list(df_quezhen)
date = []
for string in date_list:
    length = len(string)
    string = string[0:length-8]
    date.append(string)
del date[0]

districts = df['address'].tolist()
```

## 每日新增确诊

```python
line_quezhen = pyc.Line()
line_quezhen.add_xaxis(date)
for district in districts:
    index = df_quezhen[(df_quezhen['address']==district)].index.tolist()[0]
    data_list = df_quezhen.values.tolist()[index]
    del data_list[0]
    line_quezhen.add_yaxis(district,data_list)
line_quezhen.render_notebook()
```

## 每日新增无症状感染者

```python
line_wuzhengzhuang = pyc.Line()
line_wuzhengzhuang.add_xaxis(date)
for district in districts:
    index = df_wuzhengzhuang[(df_wuzhengzhuang['address']==district)].index.tolist()[0]
    data_list = df_wuzhengzhuang.values.tolist()[index]
    del data_list[0]
    line_wuzhengzhuang.add_yaxis(district,data_list)
line_wuzhengzhuang.render_notebook()
```
