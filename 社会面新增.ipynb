```python
from matplotlib import pyplot as plt
from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pyecharts.charts as pyc

plt.rcParams['font.sans-serif'] = [u'SimHei']
plt.rcParams['axes.unicode_minus'] = False

data1 = pd.read_csv("social_screening_quezhen.csv",index_col=0)
data2 = pd.read_csv("social_screening_wuzhengzhuang.csv",index_col=0)
data3 = pd.read_csv("Shanghai.csv",index_col=0)

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df3 = pd.DataFrame(data3)

df1.plot(kind="line")
plt.title("上海市各区社会面确诊人数")
df2.plot(kind="line")
plt.title("上海市各区社会面无症状感染人数")
df3.plot(kind="line")
plt.title("上海市全市社会面确诊及无症状感染人数")

plt.show()
```
