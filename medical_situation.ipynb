- 各区进入医院的人数比例（结合各区总人数）【搁置】
- 上海方舱医院和普通医院的折线统计图

## 下图为各区医院处理人数的折线统计图

```python
import numpy as np
import pandas as pd
import math
from pyecharts.charts import *
from pyecharts.components import Table
from pyecharts import options as opts
from pyecharts.commons.utils import JsCode
import random
import datetime
from pyecharts.globals import CurrentConfig
CurrentConfig.ONLINE_HOST = "https://cdn.kesci.com/lib/pyecharts_assets/"
    
def getData(district, date, type):

    return \
        originData.reindex(index=[district], columns=[date + "_" + ("fangcang" if type == 'f' else "yiyuan")]).values[0][
            0]

def getHospitalData(district, date):
    temp = getData(district, date, 'y')
    return temp if not math.isnan(temp) else 0

def getFangcangData(date):
    temp = getData("Shanghai", date, 'f')
    return temp if not math.isnan(temp) else 0


originData = pd.read_csv('CaseInfo/medical_resource.csv')
originData.set_index(["address"], inplace=True)
hospitalListOfDistricts = []
for i in range(0, 16):
    hospitalListOfDistricts.append([])
namesOfDistrict = ["PudongNew", "Huangpu", "Xuhui", "Changning", "Jingan", "Putuo", "Hongkou", "Yangpu", "Minhang", "Baoshan", "Jiading", "Jinshan", "Songjiang", "Qingpu", "Fengxian", "Chongming"]

# 获取列名，添加日期
dateList = []
ShanghaiFangcangData = []
ShanghaiHospitalData = []

for eachRowName in originData.columns:
    if eachRowName[0:10] not in dateList and eachRowName != 'address':
        dateList.append(eachRowName[0:10])
        ShanghaiFangcangData.append(getFangcangData(eachRowName[0:10]))
        ShanghaiHospitalData.append(getHospitalData("Shanghai", eachRowName[0:10]))
        for eachDistrict in range(0,16):
            hospitalListOfDistricts[eachDistrict].append(getHospitalData(namesOfDistrict[eachDistrict], eachRowName[0:10]))
        
    # pudongnewHospitalList.append(getHospitalData("PudongNew", eachRowName[0:10]))

line = (Line()
        .add_xaxis(dateList)
        .add_yaxis(namesOfDistrict[0], hospitalListOfDistricts[0])
        .add_yaxis(namesOfDistrict[1], hospitalListOfDistricts[1])
        .add_yaxis(namesOfDistrict[2], hospitalListOfDistricts[2])
        .add_yaxis(namesOfDistrict[3], hospitalListOfDistricts[3])
        .add_yaxis(namesOfDistrict[4], hospitalListOfDistricts[4])
        .add_yaxis(namesOfDistrict[5], hospitalListOfDistricts[5])
        .add_yaxis(namesOfDistrict[6], hospitalListOfDistricts[6])
        .add_yaxis(namesOfDistrict[7], hospitalListOfDistricts[7])
        .add_yaxis(namesOfDistrict[8], hospitalListOfDistricts[8])
        .add_yaxis(namesOfDistrict[9], hospitalListOfDistricts[9])
        .add_yaxis(namesOfDistrict[10], hospitalListOfDistricts[10])
        .add_yaxis(namesOfDistrict[11], hospitalListOfDistricts[11])
        .add_yaxis(namesOfDistrict[12], hospitalListOfDistricts[12])
        .add_yaxis(namesOfDistrict[13], hospitalListOfDistricts[13])
        .add_yaxis(namesOfDistrict[14], hospitalListOfDistricts[14])
        .add_yaxis(namesOfDistrict[15], hospitalListOfDistricts[15])
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(
            datazoom_opts=[opts.DataZoomOpts(), opts.DataZoomOpts(type_="inside")],
        )
       )

lineTotal = (Line()
            .add_xaxis(dateList)
            .add_yaxis("上海方舱医院人数", ShanghaiFangcangData)
            .add_yaxis("上海医院人数", ShanghaiHospitalData)
            .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(
            datazoom_opts=[opts.DataZoomOpts(), opts.DataZoomOpts(type_="inside")],
        )
            )
lineTotal.render_notebook()


```
