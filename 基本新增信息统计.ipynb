```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import pyecharts.charts as pyc
from pyecharts.components import Table
from pyecharts import options as opts
from pyecharts.commons.utils import JsCode
import random
import datetime
from pyecharts.globals import CurrentConfig
CurrentConfig.ONLINE_HOST = "https://cdn.kesci.com/lib/pyecharts_assets/"
```

```python
data = pd.read_csv('CaseInfo/daily_report_by_address.csv')

df = pd.DataFrame(data)
```

```python
df.head()

df.tail()
```

```python
df.loc['col_sum'] = df.apply(lambda x:x.sum())
```

```python
daily_sum = df.loc['col_sum']
daily_sum_df = pd.DataFrame(daily_sum)
daily_sum_df.columns = ['sum']
daily_sum_df[['sum']] = daily_sum_df[['sum']].astype('int')
```

```python
date = list(daily_sum_df.index)
num = daily_sum_df['sum'].tolist()
del date[0]
del num[0]
line = (pyc.Line()
       .add_xaxis(date)
       .add_yaxis('每日新增数量',num)
       .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(
            datazoom_opts=[opts.DataZoomOpts(), opts.DataZoomOpts(type_="inside")],
        )
       )
line.render_notebook()
```
